```python
# I've took the solution from this [site](https://pythonnumericalmethods.berkeley.edu/notebooks/chapter22.03-The-Euler-Method.html) and tried to complete exercise
import numpy as np
def euler_method(position, velocity, acceleration, steps, delta):
  table = {key: [] for key in ['s(t)', 'v(t)', 'a(t)']}
  h = delta # Step size h = 0.01
  t = np.arange(0, steps * h, h) # Numerical grid np.arange(0, 1 + h, h)
  s0 = position # -1
  a = acceleration

  # Explicit Euler Method
  s = list(((0,0),)*len(t)) # np.zeros(len(t))
  s[0] = s0
  v = list(((0,0),)*len(t))
  v[0] = velocity

  for i in range(0, len(t)-1):
    tmp = (v[i][0] + a[0] * t[i], v[i][1] + a[1] * t[i])
    v[i+1] = tmp
    table['v(t)'].append(v[i])
    tmp = (s[i][0] + v[i][0] * t[i], s[i][1] + v[i][1] * t[i])
    s[i+1] = tmp
    table['s(t)'].append(s[i])
  table['a(t)'] = list((a,)*len(t))
  return table
```

```python
print(euler_method((0,0), (1,0), (0,0.2), 6, 2))
```
