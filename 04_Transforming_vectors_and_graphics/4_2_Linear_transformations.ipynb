```python
# Exercise 4.16
def add(*vectors): 
 return (sum([v[0] for v in vectors]), sum([v[1] for v in vectors]))

v1 = (5,2)
v2 = (3,4)
v3 = add(v1,v2)
draw(Arrow(v1), Arrow(v2), Arrow(v3))
```

```python
def reflec_vec_x(vector):
  return (vector[0], vector[1]*(-1))

v1 = reflec_vec_x(v1)
v2 = reflec_vec_x(v2)
v3 = reflec_vec_x(v3)
draw(Arrow(v1), Arrow(v2), Arrow(v3))
```

```python
def scale(scalar,v):
    return (scalar * v[0], scalar * v[1])

draw(Arrow(scale(2,v3)), Arrow(scale(2,reflec_vec_x(v3))))
```
