```python
# Exercise 2.25
# Solutions from the book
def distance(v1,v2):
  return length(subtract(v1,v2))

def perimeter(vectors):
  distances = [distance(vectors[i], vectors[(i+1)%len(vectors)]) for i in range(0,len(vectors))]
  return sum(distances)
```
