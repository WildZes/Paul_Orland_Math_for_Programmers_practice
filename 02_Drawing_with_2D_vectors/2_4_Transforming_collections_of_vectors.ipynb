```python
# Exercise 2.4
from math import sin, cos, atan2, sqrt, tan, pi
from vector_drawing import *

dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),   (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3), (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)]

def add(*vectors): 
 return (sum([v[0] for v in vectors]), sum([v[1] for v in vectors]))
 
def translate(translation, vectors):
  return [add(translation, v) for v in vectors]

def subtract(v1,v2):
  return (v1[0]-v2[0], v1[1]-v2[1])

def distance(v1,v2):
  return length(subtract(v1,v2))

def length(v):
  return sqrt(v[0]**2 + v[1]**2)

def to_cartesian(polar_vector):
  length, angle = polar_vector[0], polar_vector[1]
  return (length*cos(angle), length*sin(angle))

def to_polar(vector):
  x, y = vector[0], vector[1]
  angle = atan2(y,x)
  return (length(vector), angle)

def rotate(angle, vectors):
  vectors = [to_polar(v) for v in vectors]
  vectors = [to_cartesian((v[0], v[1]+angle)) for v in vectors]
  return vectors
```

```python
# Solution from the book
def rotateII(angle, vectors):
  polars = [to_polar(v) for v in vectors]
  return [to_cartesian((l, a+angle)) for l,a in polars]
```

```python
draw(Polygon(*rotateII(pi/9,dino_vectors)))
```

```python
draw(Polygon(*rotate(pi/9,dino_vectors)))
```

```python
# Exercise 2.43
def regular_polygon(n):
  angle = 2*pi/n
  s = to_polar((1,0))
  vectors = [to_cartesian((s[0], s[1]+angle*i))for i in range(n+1)]
  draw(Polygon(*vectors))
```

```python
regular_polygon(10)
```
